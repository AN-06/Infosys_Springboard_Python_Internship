```python
#task 1- calculator with oops concept
class Operation:
    def execute(self, a, b):
        pass


class Add(Operation):
    def execute(self, a, b):
        return a + b


class Sub(Operation):
    def execute(self, a, b):
        return a - b


class Mul(Operation):
    def execute(self, a, b):
        return a * b


class Div(Operation):
    def execute(self, a, b):
        if b == 0:
            return "Division by zero not allowed"
        return a / b


class Calculator:
    def __init__(self):
        self.operations = {
            "+": Add(),
            "-": Sub(),
            "*": Mul(),
            "/": Div()
        }

    def calculate(self, a, b, op):
        if op in self.operations:
            return self.operations[op].execute(a, b)
        else:
            return "Invalid operation"

calc = Calculator()
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operation (+,-,*,/): ")

print("Result:", calc.calculate(a, b, op))
```

```python
#task 2- multiple, multilevel, hierarchical, hybrid inheritance
# Base Class
class A:
    def showA(self):
        print("Class A")

#Multilevel
class B(A):   # B inherits A
    def showB(self):
        print("Class B")

class C(B):
    def showC(self):
        print("Class C")

#Hierarchical
class D(A):
    def showD(self):
        print("Class D")

#Multiple & Hybrid
class E(B, D):
    def showE(self):
        print("Class E")



objC = C()   # Multilevel
print("\nMultilevel Inheritance")
objC.showA()
objC.showB()
objC.showC()

objD = D()   # Hierarchical
print("\nHierarchical Inheritance")
objD.showA()
objD.showD()

objE = E()   # Multiple + Hybrid
print("\nMultiple & Hybrid Inheritance")
objE.showA()
objE.showB()
objE.showD()
objE.showE()
```

```python
#task 3 - code for over riding and overloading
#overriding
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

obj = Dog()
obj.sound()
```

```python
#overloading
class Math:
    def add(self, a=0, b=0, c=0):
        return a + b + c

obj = Math()
print(obj.add(5, 10))
print(obj.add(2, 3, 4))
print(obj.add(7))
```
